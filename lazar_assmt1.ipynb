# 5. Perform meaningful visualizations using the database

# Histograms of numerical columns

```python
plt.figure(figsize=(10,6))
plt.subplot(2,2,1)

plt.hist(df['SL'],bins=10,edgecolor='y',color='g')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
#plt.title('Histogram of Sepal Length')

plt.subplot(2,2,2)
plt.hist(df['SW'],bins=10,edgecolor='y',color='g')
plt.xlabel('Sepal Width')
plt.ylabel('Frequency')
#plt.title('Histogram of Sepal Width')

plt.subplot(2,2,3)
plt.hist(df['PL'],bins=10,edgecolor='y',color='g')
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
#plt.title('Histogram of Petal Length')

plt.subplot(2,2,4)
plt.hist(df['PW'],bins=10,edgecolor='y',color='g')
plt.xlabel('Petal Width')
plt.ylabel('Frequency')
plt.show()
```

### Histograms shown above suggests that SL, SW and PL column values follow approximately normal distribution, whereas, PW column values do not follow normal distribution.

## Box Plots

```python
plt.figure(figsize=(10,6))

plt.subplot(2,2,1)
sns.boxplot(x = 'Classification', y='SL', data = df)

plt.subplot(2,2,2)
sns.boxplot(x = 'Classification', y='SW', data = df)

plt.subplot(2,2,3)
sns.boxplot(x = 'Classification', y='PL', data = df)


plt.subplot(2,2,4)
sns.boxplot(x = 'Classification', y='PW', data = df)
plt.show()
```

## Correlation Plot

```python
sns.heatmap(data=df.corr(),annot=True)
```

### Correlation plot shows that Length and width of Sepal of Iris flower Sepal have strong relationship with the lenth and width of their Petal
